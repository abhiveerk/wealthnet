```python
import pandas as pd
from datetime import datetime, timedelta

from google.colab import drive
drive.mount('/content/drive')

nifty = pd.read_csv('/content/drive/My Drive/FINALNIFTYWITHTI.csv')

# Adjust the paths as needed
nifty = pd.read_csv('/content/drive/My Drive/FINALNIFTYWITHTI.csv')
midcap = pd.read_csv('/content/drive/My Drive/FINALMIDCPNIFTYWITHTI.csv')
bank = pd.read_csv('/content/drive/My Drive/FINALBANKNIFTYWITHTI.csv')
fin = pd.read_csv('/content/drive/My Drive/FINALFINNIFTYWITHTI.csv')

def prepare_data(df):
    df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
    df.set_index('TIMESTAMP', inplace=True)
    return df

nifty = prepare_data(nifty)
midcap = prepare_data(midcap)
bank = prepare_data(bank)
fin = prepare_data(fin)

nifty['DIR_previous_nifty'] = nifty['DIR'].shift(1)
midcap['DIR_previous_midcap'] = midcap['DIR'].shift(1)
bank['DIR_previous_bank'] = bank['DIR'].shift(1)
fin['DIR_previous_fin'] = fin['DIR'].shift(1)

# Merge based on the index
merged_df_nifty = nifty.join([midcap['DIR_previous_midcap'], bank['DIR_previous_bank'], fin['DIR_previous_fin']])
merged_df_nifty = merged_df_nifty.drop('DIR_previous_nifty', axis=1)

merged_df_midcap = midcap.join([nifty['DIR_previous_nifty'], bank['DIR_previous_bank'], fin['DIR_previous_fin']])
merged_df_midcap= merged_df_midcap.drop('DIR_previous_midcap', axis=1)

merged_df_bank = bank.join([midcap['DIR_previous_midcap'], nifty['DIR_previous_nifty'], fin['DIR_previous_fin']])
merged_df_bank = merged_df_bank.drop('DIR_previous_bank', axis=1)

merged_df_fin = fin.join([midcap['DIR_previous_midcap'], bank['DIR_previous_bank'], nifty['DIR_previous_nifty']])
merged_df_fin = merged_df_fin.drop('DIR_previous_fin', axis=1)

merged_df_nifty.to_csv('/content/drive/My Drive/updated_NIFTY.csv', index=False)
# updated_nifty = pd.read_csv('/content/drive/My Drive/updated_NIFTY.csv')

merged_df_midcap.to_csv('/content/drive/My Drive/updated_MIDCAP.csv', index=False)
merged_df_bank.to_csv('/content/drive/My Drive/updated_BANK.csv', index=False)
merged_df_fin.to_csv('/content/drive/My Drive/updated_FIN.csv', index=False)



nifty.head(5)

```
